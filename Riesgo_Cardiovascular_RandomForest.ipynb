```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- 1. Load Data (Existing code) ---
from google.colab import drive
drive.mount('/content/drive')

ruta = "/content/drive/MyDrive/Maestria IA en Salud/Modulo 2/heart_disease.csv"
df = pd.read_csv(ruta)

print("Primeras filas del dataset:")
display(df.head())

print("\nResumen de columnas, tipos de datos y valores nulos:")
df.info()

print("\nEstad√≠sticas descriptivas:")
display(df.describe())

print("\nValores nulos por columna:")
display(df.isnull().sum())

# Data distribution and correlation matrix (as in your notebook)
# ... (Your plotting code for TenYearCHD, age, sysBP, cigsPerDay, totChol, currentSmoker)
# ... (Your correlation matrix plotting code)

# --- 2. Data Preprocessing and Feature Engineering (Enhanced) ---
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_classif

columna_objetivo = "TenYearCHD"
columna_eliminar = ["education"] # Keep this as per your analysis

# Create a copy to avoid modifying the original df for raw data inspection
df_processed = df.copy()

# --- Feature Engineering Step 1: Handle specific NaN logic for cigsPerDay ---
# If currentSmoker is 0 (non-smoker) and cigsPerDay is NaN, it's logically 0.
df_processed.loc[(df_processed['currentSmoker'] == 0) & (df_processed['cigsPerDay'].isnull()), 'cigsPerDay'] = 0

# --- Drop 'education' column ---
df_processed = df_processed.drop(columns=[c for c in columna_eliminar if c in df_processed.columns])

# Separate features (X) and target (y)
X = df_processed.drop(columns=[columna_objetivo]).copy()
y = df_processed[columna_objetivo].copy()

# --- Split data FIRST to avoid data leakage ---
X_train_raw, X_val_raw, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# --- Imputation (fit only on training data, transform both) ---
imputer = SimpleImputer(strategy="median") # Median is good for outliers
X_train_imputed = imputer.fit_transform(X_train_raw)
X_val_imputed = imputer.transform(X_val_raw)

# Convert back to DataFrame (optional, but good for readability/debugging)
X_train_imputed = pd.DataFrame(X_train_imputed, columns=X_train_raw.columns, index=X_train_raw.index)
X_val_imputed = pd.DataFrame(X_val_imputed, columns=X_val_raw.columns, index=X_val_raw.index)

print("\nConfirmed no missing values after imputation (Training Set):")
print(X_train_imputed.isnull().sum())
print("\nConfirmed no missing values after imputation (Validation Set):")
print(X_val_imputed.isnull().sum())

# --- Feature Engineering Step 2 (Interaction terms, polynomial features) ---
# Example interaction terms (choose based on domain knowledge or further correlation analysis)
X_train_fe = X_train_imputed.copy()
X_val_fe = X_val_imputed.copy()

X_train_fe['age_x_sysBP'] = X_train_fe['age'] * X_train_fe['sysBP']
X_val_fe['age_x_sysBP'] = X_val_fe['age'] * X_val_fe['sysBP']

X_train_fe['BMI_x_diabetes'] = X_train_fe['BMI'] * X_train_fe['diabetes']
X_val_fe['BMI_x_diabetes'] = X_val_fe['BMI'] * X_val_fe['diabetes']

# --- Scaling (fit only on training data, transform both) ---
# Using StandardScaler as it's common; consider RobustScaler if outliers are a severe problem after imputation
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_fe)
X_val_scaled = scaler.transform(X_val_fe)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train_fe.columns, index=X_train_fe.index)
X_val_scaled = pd.DataFrame(X_val_scaled, columns=X_val_fe.columns, index=X_val_fe.index)


# --- 3. Advanced Feature Selection ---

# --- Option 1: PCA (as you used, but now correctly applied after split) ---
print("\n--- PCA for Dimensionality Reduction ---")
pca_full = PCA().fit(X_train_scaled) # Fit PCA only on training data
varianza = pca_full.explained_variance_ratio_
varianza_explicada = np.cumsum(varianza)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(varianza_explicada)+1), varianza_explicada, marker="o")
plt.axhline(0.95, linestyle="--", color="r", label="95% Variance Explained")
plt.xlabel("N√∫mero de componentes")
plt.ylabel("Varianza explicada acumulada")
plt.title("Selecci√≥n de caracter√≠sticas con PCA")
plt.legend()
plt.grid(True)
plt.show()

n_components = int(np.searchsorted(varianza_explicada, 0.95) + 1)
print(f"Number of components for 95% variance: {n_components}")

pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train_scaled) # Fit and transform training
X_val_pca = pca.transform(X_val_scaled) # Only transform validation

X_train_final = pd.DataFrame(X_train_pca, columns=[f"PC{i}" for i in range(1, n_components+1)], index=X_train_scaled.index)
X_val_final = pd.DataFrame(X_val_pca, columns=[f"PC{i}" for i in range(1, n_components+1)], index=X_val_scaled.index)

print(f"Shape of training features after PCA: {X_train_final.shape}")
print(f"Shape of validation features after PCA: {X_val_final.shape}")

# --- Option 2: SelectKBest (as an alternative or to combine before PCA) ---
# from sklearn.feature_selection import SelectKBest, f_classif
# print("\n--- SelectKBest for Feature Selection ---")
# selector = SelectKBest(f_classif, k=10) # Example: select top 10 features
# X_train_selected = selector.fit_transform(X_train_scaled, y_train)
# X_val_selected = selector.transform(X_val_scaled)

# selected_features_names = X_train_scaled.columns[selector.get_support()]
# print(f"Selected features: {list(selected_features_names)}")

# X_train_final = pd.DataFrame(X_train_selected, columns=selected_features_names, index=X_train_scaled.index)
# X_val_final = pd.DataFrame(X_val_selected, columns=selected_features_names, index=X_val_scaled.index)


# --- 4. Model Training and Advanced Hyperparameter Tuning ---
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier # For MLP
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, classification_report, confusion_matrix
import warnings

warnings.filterwarnings('ignore') # Suppress warnings during grid search for cleaner output

UMBRAL = 0.3 # Keep this as your desired threshold for high recall

# Define a common StratifiedKFold for cross-validation
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Models and their hyperparameter grids for tuning
# 'class_weight="balanced"' is included for models that support it
modelos_tuned = {
    "LogisticRegression": {
        'estimator': LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42),
        'param_grid': {'C': [0.001, 0.01, 0.1, 1, 10, 100]},
        'search_type': 'GridSearchCV'
    },
    "RandomForest": {
        'estimator': RandomForestClassifier(class_weight="balanced", random_state=42),
        'param_grid': {
            'n_estimators': [100, 200, 300],
            'max_depth': [5, 10, 15],
            'min_samples_split': [2, 5],
            'min_samples_leaf': [1, 2]
        },
        'search_type': 'RandomizedSearchCV' # Use RandomizedSearchCV for larger grids
    },
    "SVM (linear)": {
        'estimator': SVC(kernel="linear", probability=True, class_weight="balanced", random_state=42),
        'param_grid': {'C': [0.1, 1, 10, 100]},
        'search_type': 'GridSearchCV'
    },
    "GradientBoosting": {
        'estimator': GradientBoostingClassifier(random_state=42),
        'param_grid': {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.1, 0.2],
            'max_depth': [3, 5]
        },
        'search_type': 'GridSearchCV'
    },
    "KNN": { # KNN doesn't have class_weight directly; consider imblearn pipelines if this is a top performer
        'estimator': KNeighborsClassifier(),
        'param_grid': {'n_neighbors': [3, 5, 7, 9, 11], 'weights': ['uniform', 'distance']},
        'search_type': 'GridSearchCV'
    },
    "MLPClassifier": { # Added MLP for exploration
        'estimator': MLPClassifier(random_state=42, max_iter=1000),
        'param_grid': {
            'hidden_layer_sizes': [(50,), (100,), (50, 50)],
            'activation': ['relu', 'tanh'],
            'solver': ['adam'],
            'alpha': [0.0001, 0.001] # L2 regularization
        },
        'search_type': 'RandomizedSearchCV' # MLP can have a large search space
    }
}

tuned_results = []
roc_curves = []

for nombre, config in modelos_tuned.items():
    print(f"\n--- Tuning {nombre} ---")
    estimator = config['estimator']
    param_grid = config['param_grid']
    search_type = config['search_type']

    if search_type == 'GridSearchCV':
        search_model = GridSearchCV(estimator, param_grid, cv=cv_strategy, scoring='recall', n_jobs=-1, verbose=1)
    elif search_type == 'RandomizedSearchCV':
        search_model = RandomizedSearchCV(estimator, param_grid, cv=cv_strategy, scoring='recall', n_jobs=-1, verbose=1, n_iter=20, random_state=42) # n_iter for samples

    search_model.fit(X_train_final, y_train) # Use the PCA-transformed data

    best_clf = search_model.best_estimator_
    print(f"Best parameters for {nombre}: {search_model.best_params_}")
    print(f"Best cross-validation score (recall): {search_model.best_score_:.4f}")

    # Evaluate on the validation set
    y_prob = best_clf.predict_proba(X_val_final)[:, 1]
    y_pred = (y_prob >= UMBRAL).astype(int)

    tuned_results.append({
        "Modelo": nombre,
        "Best Params": search_model.best_params_,
        "Accuracy": accuracy_score(y_val, y_pred),
        "Precision": precision_score(y_val, y_pred, zero_division=0),
        "Recall": recall_score(y_val, y_pred, zero_division=0),
        "F1": f1_score(y_val, y_pred, zero_division=0),
        "AUC": roc_auc_score(y_val, y_prob),
        "Positivos predichos": int(y_pred.sum())
    })

    fpr, tpr, _ = roc_curve(y_val, y_prob)
    auc_val = roc_auc_score(y_val, y_prob)
    roc_curves.append((nombre, fpr, tpr, auc_val))

# --- 5. Comparativa de Modelos (After Tuning) ---
df_tuned_algos = pd.DataFrame(tuned_results).sort_values(by=["Recall", "AUC"], ascending=[False, False]).reset_index(drop=True)
print("\n--- Comparativa de algoritmos despu√©s de Tuning (umbral com√∫n =", UMBRAL, ") ---")
display(df_tuned_algos)

plt.figure(figsize=(10, 7))
for nombre, fpr, tpr, auc_val in roc_curves:
    plt.plot(fpr, tpr, label=f"{nombre} (AUC={auc_val:.3f})")
plt.plot([0,1],[0,1],'--', color='gray', alpha=0.7)
plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR) - Recall")
plt.title("Curvas ROC - Comparativa de algoritmos despu√©s de Tuning")
plt.legend()
plt.grid(True)
plt.show()


# --- 6. Detailed Evaluation of the Best Model ---
best_model_info = df_tuned_algos.loc[0]
best_model_name = best_model_info['Modelo']

# Retrieve the best estimator from the tuning process for detailed evaluation
# This requires rerunning the specific search for that model if you haven't stored it
print(f"\n--- Detailed Evaluation for the best model: {best_model_name} ---")
final_clf_config = modelos_tuned[best_model_name]
final_estimator = final_clf_config['estimator']
final_param_grid = final_clf_config['param_grid']
final_search_type = final_clf_config['search_type']

if final_search_type == 'GridSearchCV':
    final_search_model = GridSearchCV(final_estimator, final_param_grid, cv=cv_strategy, scoring='recall', n_jobs=-1, verbose=0)
elif final_search_type == 'RandomizedSearchCV':
    final_search_model = RandomizedSearchCV(final_estimator, final_param_grid, cv=cv_strategy, scoring='recall', n_jobs=-1, verbose=0, n_iter=20, random_state=42)

final_search_model.fit(X_train_final, y_train)
final_model = final_search_model.best_estimator_

y_prob_final = final_model.predict_proba(X_val_final)[:, 1]
y_pred_final = (y_prob_final >= UMBRAL).astype(int)

print(f"Best parameters: {final_search_model.best_params_}")
print("\nReporte de clasificaci√≥n con el modelo final (Validation Set):")
print(classification_report(y_val, y_pred_final, digits=3))

mc_final = confusion_matrix(y_val, y_pred_final)
plt.figure(figsize=(6, 5))
sns.heatmap(mc_final, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title(f"Matriz de confusi√≥n - {best_model_name} (umbral={UMBRAL})")
plt.xlabel("Predicho")
plt.ylabel("Real")
plt.show()

fpr_final, tpr_final, _ = roc_curve(y_val, y_prob_final)
auc_final = roc_auc_score(y_val, y_prob_final)
plt.figure(figsize=(6, 5))
plt.plot(fpr_final, tpr_final, label=f"{best_model_name} (AUC={auc_final:.3f})")
plt.plot([0,1],[0,1],'--', color='gray', alpha=0.7)
plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR) - Recall")
plt.title("Curva ROC - Modelo Final")
plt.legend()
plt.grid(True)
plt.show()
```
