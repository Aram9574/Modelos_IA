# 4. Preprocesamiento de datos

```python
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_classif

columna_objetivo = "TenYearCHD"
columnas_eliminar = ["education"] # Con la matriz de correlacion se puede descartar

# Crear una copia para evitar modificar el df original para inspecci√≥n de datos crudos
df_procesado = df.copy()

# Paso 1 de Ingenier√≠a de Caracter√≠sticas: Manejar l√≥gica espec√≠fica de NaN para cigsPerDay
# Si currentSmoker es 0 (no fumador) y cigsPerDay es NaN, l√≥gicamente es 0.
df_procesado.loc[(df_procesado['currentSmoker'] == 0) & (df_procesado['cigsPerDay'].isnull()), 'cigsPerDay'] = 0

# Eliminar columna seleccionada
df_procesado = df_procesado.drop(columns=[c for c in columnas_eliminar if c in df_procesado.columns])

# Separar caracter√≠sticas (X) y objetivo (y)
X = df_procesado.drop(columns=[columna_objetivo]).copy()
y = df_procesado[columna_objetivo].copy()

# Dividir los datos PRIMERO para evitar fuga de datos (Data Leakage)
X_entrenamiento_crudo, X_validacion_crudo, y_entrenamiento, y_validacion = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Imputaci√≥n (ajustar (fit) solo a los datos de entrenamiento, transformar ambos)
imputador = SimpleImputer(strategy="median") # La mediana es buena para valores at√≠picos
X_entrenamiento_imputado = imputador.fit_transform(X_entrenamiento_crudo)
X_validacion_imputado = imputador.transform(X_validacion_crudo)

# Convertir de vuelta a DataFrame
X_entrenamiento_imputado = pd.DataFrame(X_entrenamiento_imputado, columns=X_entrenamiento_crudo.columns, index=X_entrenamiento_crudo.index)
X_validacion_imputado = pd.DataFrame(X_validacion_imputado, columns=X_validacion_crudo.columns, index=X_validacion_crudo.index)

print("\nConfirmado, no hay valores perdidos despu√©s de la imputaci√≥n (Conjunto de Entrenamiento):")
print(X_entrenamiento_imputado.isnull().sum())
print("\nConfirmado, no hay valores perdidos despu√©s de la imputaci√≥n (Conjunto de Validaci√≥n):")
print(X_validacion_imputado.isnull().sum())


# Paso 2 de Ingenier√≠a de Caracter√≠sticas (T√©rminos de interacci√≥n, caracter√≠sticas polin√≥micas)
# Ejemplo de t√©rminos de interacci√≥n (elegir basado en conocimiento del dominio o an√°lisis de correlaci√≥n adicional)
X_entrenamiento_fe = X_entrenamiento_imputado.copy()
X_validacion_fe = X_validacion_imputado.copy()

X_entrenamiento_fe['edad_x_sysBP'] = X_entrenamiento_fe['age'] * X_entrenamiento_fe['sysBP']
X_validacion_fe['edad_x_sysBP'] = X_validacion_fe['age'] * X_validacion_fe['sysBP']

X_entrenamiento_fe['IMC_x_diabetes'] = X_entrenamiento_fe['BMI'] * X_entrenamiento_fe['diabetes']
X_validacion_fe['IMC_x_diabetes'] = X_validacion_fe['BMI'] * X_validacion_fe['diabetes']

# Escalar (ajustar solo a los datos de entrenamiento, transformar ambos)
# Usando StandardScaler como es com√∫n; considerar RobustScaler si los valores at√≠picos son un problema severo despu√©s de la imputaci√≥n
escalador = StandardScaler()
X_entrenamiento_escalado = escalador.fit_transform(X_entrenamiento_fe)
X_validacion_escalado = escalador.transform(X_validacion_fe)

X_entrenamiento_escalado = pd.DataFrame(X_entrenamiento_escalado, columns=X_entrenamiento_fe.columns, index=X_entrenamiento_fe.index)
X_validacion_escalado = pd.DataFrame(X_validacion_escalado, columns=X_validacion_fe.columns, index=X_validacion_fe.index)
```

# 5. Seleccion de caracteristicas

```python
# Selecci√≥n Avanzada de Caracter√≠sticas
"""Se puede usar tanto PCA como KBest, probar con cada una a ver si los resultados
cambian, dejare comentada la opcion 2"""

# Opci√≥n 1: PCA
print("\n--- PCA para Reducci√≥n de Dimensionalidad ---")
pca_completo = PCA().fit(X_entrenamiento_escalado) # Ajustar PCA solo a los datos de entrenamiento
varianza = pca_completo.explained_variance_ratio_
varianza_acumulada_explicada = np.cumsum(varianza)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(varianza_acumulada_explicada)+1), varianza_acumulada_explicada, marker="o")
plt.axhline(0.95, linestyle="--", color="r", label="95% de Varianza Explicada")
plt.xlabel("N√∫mero de componentes")
plt.ylabel("Varianza explicada acumulada")
plt.title("Selecci√≥n de caracter√≠sticas con PCA")
plt.legend()
plt.grid(True)
plt.show()

n_componentes = int(np.searchsorted(varianza_acumulada_explicada, 0.95) + 1)
print(f"N√∫mero de componentes para el 95% de varianza: {n_componentes}")

pca = PCA(n_components=n_componentes)
X_entrenamiento_pca = pca.fit_transform(X_entrenamiento_escalado) # Ajustar y transformar entrenamiento
X_validacion_pca = pca.transform(X_validacion_escalado) # Solo transformar validaci√≥n

X_entrenamiento_final = pd.DataFrame(X_entrenamiento_pca, columns=[f"PC{i}" for i in range(1, n_componentes+1)], index=X_entrenamiento_escalado.index)
X_validacion_final = pd.DataFrame(X_validacion_pca, columns=[f"PC{i}" for i in range(1, n_componentes+1)], index=X_validacion_escalado.index)

print(f"Forma de las caracter√≠sticas de entrenamiento despu√©s de PCA: {X_entrenamiento_final.shape}")
print(f"Forma de las caracter√≠sticas de validaci√≥n despu√©s de PCA: {X_validacion_final.shape}")

# --- Opci√≥n 2: SelectKBest (como alternativa o para combinar antes de PCA) ---
# from sklearn.feature_selection import SelectKBest, f_classif
# print("\n--- SelectKBest para Selecci√≥n de Caracter√≠sticas ---")
# selector = SelectKBest(f_classif, k=10) # Ejemplo: seleccionar las 10 mejores caracter√≠sticas
# X_entrenamiento_seleccionado = selector.fit_transform(X_entrenamiento_escalado, y_entrenamiento)
# X_validacion_seleccionado = selector.transform(X_validacion_escalado)

# nombres_caracteristicas_seleccionadas = X_entrenamiento_escalado.columns[selector.get_support()]
# print(f"Caracter√≠sticas seleccionadas: {list(nombres_caracteristicas_seleccionadas)}")

# X_entrenamiento_final = pd.DataFrame(X_entrenamiento_seleccionado, columns=nombres_caracteristicas_seleccionadas, index=X_entrenamiento_escalado.index)
# X_validacion_final = pd.DataFrame(X_validacion_seleccionado, columns=nombres_caracteristicas_seleccionadas, index=X_validacion_escalado.index)
```

# 6. Entrenamiento de modelos y ajuste de hiperparametros

```python

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier # A ver si MLP sirve para este caso
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, classification_report, confusion_matrix
import warnings

warnings.filterwarnings('ignore') # Suprimir advertencias durante la b√∫squeda en cuadr√≠cula para una salida limpia

UMBRAL = 0.3 # Priorizando el recall

# Definir estrategia de StratifiedKFold para validaci√≥n cruzada
estrategia_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Modelos y sus grillas de hiperpar√°metros para ajuste
# 'class_weight="balanced"' est√° incluido para modelos que lo soportan
modelos_ajustados = {
    "LogisticRegression": {
        'estimator': LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42),
        'param_grid': {'C': [0.001, 0.01, 0.1, 1, 10, 100]},
        'search_type': 'GridSearchCV'
    },
    "RandomForest": {
        'estimator': RandomForestClassifier(class_weight="balanced", random_state=42),
        'param_grid': {
            'n_estimators': [100, 200, 300],
            'max_depth': [5, 10, 15],
            'min_samples_split': [2, 5],
            'min_samples_leaf': [1, 2]
        },
        'search_type': 'RandomizedSearchCV' # Usar RandomizedSearchCV para grillas m√°s grandes
    },
    "SVM (linear)": {
        'estimator': SVC(kernel="linear", probability=True, class_weight="balanced", random_state=42),
        'param_grid': {'C': [0.1, 1, 10, 100]},
        'search_type': 'GridSearchCV'
    },
    "GradientBoosting": {
        'estimator': GradientBoostingClassifier(random_state=42),
        'param_grid': {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.1, 0.2],
            'max_depth': [3, 5]
        },
        'search_type': 'GridSearchCV'
    },
    "KNN": { # KNN no tiene class_weight directamente
        'estimator': KNeighborsClassifier(),
        'param_grid': {'n_neighbors': [3, 5, 7, 9, 11], 'weights': ['uniform', 'distance']},
        'search_type': 'GridSearchCV'
    },
    "MLPClassifier": { # A√±adido MLP para exploraci√≥n
        'estimator': MLPClassifier(random_state=42, max_iter=1000),
        'param_grid': {
            'hidden_layer_sizes': [(50,), (100,), (50, 50)],
            'activation': ['relu', 'tanh'],
            'solver': ['adam'],
            'alpha': [0.0001, 0.001] # Regularizaci√≥n L2
        },
        'search_type': 'RandomizedSearchCV' # MLP puede tener un espacio de b√∫squeda grande
    }
}

resultados_ajustados = []
curvas_roc = []

for nombre, config in modelos_ajustados.items():
    print(f"\n--- Ajustando {nombre} ---")
    estimador = config['estimator']
    grilla_parametros = config['param_grid']
    tipo_busqueda = config['search_type']

    if tipo_busqueda == 'GridSearchCV':
        modelo_busqueda = GridSearchCV(estimador, grilla_parametros, cv=estrategia_cv, scoring='recall', n_jobs=-1, verbose=1)
    elif tipo_busqueda == 'RandomizedSearchCV':
        modelo_busqueda = RandomizedSearchCV(estimador, grilla_parametros, cv=estrategia_cv, scoring='recall', n_jobs=-1, verbose=1, n_iter=20, random_state=42) # n_iter para muestras

    modelo_busqueda.fit(X_entrenamiento_final, y_entrenamiento) # Usar los datos transformados por PCA

    mejor_clf = modelo_busqueda.best_estimator_
    print(f"Mejores par√°metros para {nombre}: {modelo_busqueda.best_params_}")
    print(f"Mejor puntuaci√≥n de validaci√≥n cruzada (recall): {modelo_busqueda.best_score_:.4f}")

    # Evaluar en el conjunto de validaci√≥n
    y_prob = mejor_clf.predict_proba(X_validacion_final)[:, 1]
    y_pred = (y_prob >= UMBRAL).astype(int)

    resultados_ajustados.append({
        "Modelo": nombre,
        "Mejores Par√°metros": modelo_busqueda.best_params_,
        "Accuracy": accuracy_score(y_validacion, y_pred),
        "Precision": precision_score(y_validacion, y_pred, zero_division=0),
        "Recall": recall_score(y_validacion, y_pred, zero_division=0),
        "F1": f1_score(y_validacion, y_pred, zero_division=0),
        "AUC": roc_auc_score(y_validacion, y_prob),
        "Positivos predichos": int(y_pred.sum())
    })

    fpr, tpr, _ = roc_curve(y_validacion, y_prob)
    auc_val = roc_auc_score(y_validacion, y_prob)
    curvas_roc.append((nombre, fpr, tpr, auc_val))
```

# 7. Comparativa de modelos

```python
# Despu√©s del Ajuste
df_algos_ajustados = pd.DataFrame(resultados_ajustados).sort_values(by=["Recall", "AUC"], ascending=[False, False]).reset_index(drop=True)
print("\n--- Comparativa de algoritmos despu√©s del Ajuste (umbral com√∫n =", UMBRAL, ") ---")
display(df_algos_ajustados)

plt.figure(figsize=(10, 7))
for nombre, fpr, tpr, auc_val in curvas_roc:
    plt.plot(fpr, tpr, label=f"{nombre} (AUC={auc_val:.3f})")
plt.plot([0,1],[0,1],'--', color='gray', alpha=0.7)
plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR) - Recall")
plt.title("Curvas ROC - Comparativa de algoritmos despu√©s del Ajuste")
plt.legend()
plt.grid(True)
plt.show()
```

# 8. Evaluacion del mejor modelo

```python
info_mejor_modelo = df_algos_ajustados.loc[0]
nombre_mejor_modelo = info_mejor_modelo['Modelo']

# Recuperar el mejor estimador del proceso de ajuste para una evaluaci√≥n detallada
# Esto requiere volver a ejecutar la b√∫squeda espec√≠fica para ese modelo
print(f"\n--- Evaluaci√≥n Detallada para el mejor modelo: {nombre_mejor_modelo} ---")
config_clf_final = modelos_ajustados[nombre_mejor_modelo]
estimador_final = config_clf_final['estimator']
grilla_parametros_final = config_clf_final['param_grid']
tipo_busqueda_final = config_clf_final['search_type']

if tipo_busqueda_final == 'GridSearchCV':
    modelo_busqueda_final = GridSearchCV(estimador_final, grilla_parametros_final, cv=estrategia_cv, scoring='recall', n_jobs=-1, verbose=0)
elif tipo_busqueda_final == 'RandomizedSearchCV':
    modelo_busqueda_final = RandomizedSearchCV(estimador_final, grilla_parametros_final, cv=estrategia_cv, scoring='recall', n_jobs=-1, verbose=0, n_iter=20, random_state=42)

modelo_busqueda_final.fit(X_entrenamiento_final, y_entrenamiento)
modelo_final = modelo_busqueda_final.best_estimator_

y_prob_final = modelo_final.predict_proba(X_validacion_final)[:, 1]
y_pred_final = (y_prob_final >= UMBRAL).astype(int)

print(f"Mejores par√°metros: {modelo_busqueda_final.best_params_}")
print("\nReporte de clasificaci√≥n con el modelo final (Conjunto de Validaci√≥n):")
print(classification_report(y_validacion, y_pred_final, digits=3))

mc_final = confusion_matrix(y_validacion, y_pred_final)
plt.figure(figsize=(6, 5))
sns.heatmap(mc_final, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title(f"Matriz de confusi√≥n - {nombre_mejor_modelo} (umbral={UMBRAL})")
plt.xlabel("Predicho")
plt.ylabel("Real")
plt.show()

fpr_final, tpr_final, _ = roc_curve(y_validacion, y_prob_final)
auc_final = roc_auc_score(y_validacion, y_prob_final)
plt.figure(figsize=(6, 5))
plt.plot(fpr_final, tpr_final, label=f"{nombre_mejor_modelo} (AUC={auc_final:.3f})")
plt.plot([0,1],[0,1],'--', color='gray', alpha=0.7)
plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR) - Recall")
plt.title("Curva ROC - Modelo Final")
plt.legend()
plt.grid(True)
plt.show()
```
